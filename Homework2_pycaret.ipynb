```python
import gradio as gr
from pycaret.classification import load_model, predict_model

# Load the trained model
model = load_model('best_binary_classifier')

# Define prediction function
def predict(data):
    predictions = predict_model(model, data=pd.DataFrame([data]))
    return predictions['Label'][0]

# Create Gradio interface
gradio_interface = gr.Interface(fn=predict, inputs=['number', 'number'], outputs='label')
gradio_interface.launch()
```
